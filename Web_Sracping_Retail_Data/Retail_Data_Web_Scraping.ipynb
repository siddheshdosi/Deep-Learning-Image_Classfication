```python
import re
import requests
from bs4 import BeautifulSoup
import json
import os
```

### Womens Tops

```python
# Base URL : https://www.koovs.com
site = 'https://www.koovs.com/jarvis-service/v1/product/listing/complete?href=https%3A%2F%2Fwww.koovs.com%2Fwomen%2Ftops%2F&page-size=36&sort=relevance&page='

pages = range(1,101)
IMG_DIR='./data/Tops/'
img = 1
for page in pages:
    try:
        url = site+str(page)
        response = requests.get(url)
        response_data = json.loads(response.text)
        for i in range(0,len(response_data['data'][0]['data'])):
            sub_url = response_data['data'][0]['data'][i]['imageSmallUrl']
            sub_response = requests.get(sub_url)
            filename = 'img_'+str(img)+'.jpg'
            img+=1
            IMG_PATH = os.path.join(IMG_DIR,filename)
            print('save img in : ',IMG_PATH)
            with open(IMG_PATH, 'wb') as f:
                f.write(sub_response.content)
    except:
        print('last page : ',page-1)
        break
        
    
```

### Women Sunglasses

```python
import re
import requests
from bs4 import BeautifulSoup
from random import shuffle
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
site = 'https://www.titaneyeplus.com/sunglasses-women.html?is_ajax=1&p='

pages = range(1,101)
IMG_DIR = './data/Sunglasses/'
img=1
for page in pages:
    try:
        url = site+str(page)+'&is_scroll=1'
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')
        img_tags = soup.find_all('img')
        urls = []
        for im in img_tags:
            #print(im)
            try:
                #print(im['src'])
                urls.append(im['src'])
            except:
                continue
        for url in urls:
            filename = 'img_'+str(img)+'.jpg'
            img+=1
            IMG_PATH = os.path.join(IMG_DIR,filename)
            img_flag = re.search(r'/([\w_-]+[.](jpg|gif|png|JPG|GIF|PNG))$', url)
            if not img_flag:
                img-=1
                #print("Regex didn't match with the url: {}".format(url))
                continue
            print('save img in : ',IMG_PATH)
            with open(IMG_PATH, 'wb') as f:
                if 'http' not in url:
                    # sometimes an image source can be relative 
                    # if it is provide the base url which also happens 
                    # to be the site variable atm. 
                    url = '{}{}'.format(site, url)
                response = requests.get(url)
                f.write(response.content)
        
    except:
        print('last page : ',page-1)
        break
    
```
